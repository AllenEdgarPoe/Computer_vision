```python
from matplotlib.image import imread
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

### Resizing Image using CV2

```python
img = imread("C:/Users/JKshark/Pictures/Cookie.jpg")
# print("  : "+str(img.shape))
resized_img1 = cv2.resize(img, (50,90))
resized_img2 = cv2.resize(img,(20,36))
resized_img3 = cv2.resize(img,(10,18))
hued_img = np.average(img, axis=2)

fig, (original, resized1, resized2, resized3) = plt.subplots(ncols=4, figsize = (10,5))

original.imshow(img)
resized1.imshow(resized_img1)
resized2.imshow(resized_img2)
resized3.imshow(resized_img3)


original.title.set_text(str(img.shape))
resized1.title.set_text(str(resized_img1.shape))
resized2.title.set_text(str(resized_img2.shape))
resized3.title.set_text(str(resized_img3.shape))

plt.suptitle("Resizing Image")
plt.show()
```

## Feature Extraction - Edge
Edge detect          

```python
sobelx = cv2.Sobel(img, cv2.CV_8U, 1, 0, ksize = 3)
sobely = cv2.Sobel(img, cv2.CV_8U, 0, 1, ksize = 3)
canny = cv2.Canny(img, 50,100)

fig,(x,y,can) = plt.subplots(ncols = 3, figsize = (20,10))
x.imshow(sobelx)
y.imshow(sobely)
can.imshow(canny)

fig.suptitle("Edge Detection", fontsize=40)

plt.show()
```
